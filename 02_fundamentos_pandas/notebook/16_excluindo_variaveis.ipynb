# Excluindo Variáveis do Dataframe

```python
import pandas as pd
```

```python
dados = pd.read_csv('../dados/aluguel_res_novas_variaveis.csv', sep=';')
```

```python
dados.head(10)
```

```python
dados_aux = dados[['Tipo Agregado', 'Valor Bruto', 'Valor Bruto m2', 'Valor m2', 'Tamanho']]
dados_aux.head(10)
```

## Formas de excluir variáveis

```python
# usando del
del dados_aux['Tipo Agregado']
dados_aux.head(10)
```

```python
# usando pop
dados_aux.pop('Valor Bruto')
dados_aux.head(10)
```

```python
# verificando o nome das colunas
dados.columns
```

```python
# Também é possível colacar em uma lista
list(dados.columns)
```

```python
# usando drop, é possível passar várias variáveis
# também é possível excluir linhas
# deve-se passar o eixo e o inplace para executar no df
dados.drop(['Valor Bruto', 'Valor Bruto m2', 'Tamanho'], axis=1, inplace=True)
dados.head(10)
```

```python
# exportando novamente para utilizar nos próximos notebooks
dados.to_csv('../dados/aluguel_res_del_variaveis.csv', sep=';', index=False)
```
